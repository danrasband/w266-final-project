```python
%%time

sentences = data_helper.get_sentences()
display(sentences.head())
```

```python
%%time

sentences['sentence_length'] = [len(str(row.sentence).split(' ')) for _, row in sentences.iterrows()]
display(sentences.head())
```

```python
%%time

entities = data_helper.get_entities()
entities['entity_id'] = None
display(entities.head())
```

```python
%%time

documents = data_helper.get_documents()
display(documents.head())
```

```python
documents.loc[['bn/abc/00/abc_0006@0006@abc@bn@en@on']]
```

```python
%%time

sentences['start_index'] = 0

for document_id, document in tqdm(documents.iterrows(), total=len(documents)):
    i = 0
    start_index = 0
    while True:
        try:
            sentence_id = '{}:{}'.format(document_id, i)
            sentence = sentences.loc[sentence_id]
            sentences.loc[sentence_id, 'start_index'] = start_index
            start_index = start_index + sentence['sentence_length']
            i += 1
        except KeyError:
            break

display(sentences.head())
```

```python
for entity_id, entity in tqdm(entities.iterrows(), total=len(entities)):
    sentence_id = '{}:{}'.format(entity.document_id, entity.sentence_index)
    sentence = sentences.loc[sentence_id]

    document_id = entity['document_id']
    
    start_index = sentence['start_index']
    start_word_index = start_index + entity['start_word_index']
    end_word_index = start_index + entity['end_word_index']
    
    new_entity_id = '{}:{}:{}'.format(document_id, start_word_index, end_word_index)
    
    entities.loc[entity_id, 'entity_id'] = new_entity_id
```
